```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('D:/обучение/data set/World Cities Database Population (OCT-2022).csv', sep = ';', na_filter=False )
df.head(5)
```

```python
df.isna().sum()
```

```python
df.describe ()
```

```python
# преобразовываем столбец координат
column_names = list(df.columns)
column_names[-1] = 'Coordinates'
df.columns = column_names
df['Coordinates'] = df['Coordinates'].apply(lambda x: str(x).split(',')[:2])
df['Coordinates']
```

```python
df['Latitude'] = df['Coordinates'].apply(lambda x: x[0])
df['Longitude'] = df['Coordinates'].apply(lambda x: x[-1])
```

```python
# приобразовываем столбцы в числовые
df['Latitude'] = pd.to_numeric(df['Latitude'], errors='coerce')
df['Longitude'] = pd.to_numeric(df['Longitude'], errors='coerce')
```

```python
#Namibia имеет код страны NA, который совпадает с NaN.заменить значения NaN на строку 
df['Country Code'] = df['Country Code'].fillna('NA')
```

```python
# преобразовывыем столбцы населения и высоты в числовые значения.
df['Population'] = df['Population'].apply(lambda x: pd.to_numeric(x, errors = 'coerce'))
df['Elevation'] = df['Elevation'].apply(lambda x: pd.to_numeric(x,  errors = 'coerce'))
```

```python
# удаляем лишнии стодлбцы,которые нам не понадабятся
df = df.iloc[:,[1,2,3,6,7,-9,-8,-6,-4,-2,-1]]
```

```python
df.isna().sum()
```

```python
#  в столбце Elevation много пропущенных значений, можно его удалить
df = df.drop('Elevation', axis = 1)
```

```python
# удаляем строки с пропущенными значениями
df = df.dropna()
```

```python
df.to_csv('population-cleaned.csv', sep = ',', index = False) # записываем
```
